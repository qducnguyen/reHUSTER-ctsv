# Basic EDA

```python
# dict: AId first
usr_data_clean = {}
for sample in usr_data:
    if sample["AId"] not in not_exists_event and sample["AId"] not in usr_data_clean:
        usr_data_clean[sample["AId"]] = sample

len(usr_data_clean) # 9085 - 4544
```

```python
# dict: AId first
act_data_clean = {}
for sample in act_data:
    if sample["AId"] not in not_exists_event and sample["AId"] not in act_data_clean:
        act_data_clean[sample["AId"]] = sample

len(act_data_clean) # 9085 - 4544
```

```python
from trafilatura.utils import sanitize
from trafilatura import extract

user_info  = {} # UserCode : --> Info, get only relevant fields
activity_info = {} # ACode: --> Info, get only relevant fields

data_combined = []
count = 0

for AId, sample in usr_data_clean.items():
    act_data = act_data_clean[AId]["Activities"][-1] # 8848 duplicated but it's alright now
    activity_info[AId] = {"id": AId,
                          "title": sanitize(act_data["AName"]), 
                          "type": sanitize(act_data["AType"]),
                          "desc": extract(act_data["ADesc"]),
                          "btc": act_data["GName"],
                          "startTime": act_data["StartTime"],
                          "finishTime": act_data["FinishTime"],
                          "deadlineTime": act_data["Deadline"],
                          }
    
    activity_info[AId]["criteria"] = act_data["CriteriaLst"]

    for user_data in sample["UserActivityLst"]:
        user_info[user_data['UserCode']] = {"id": user_data['UserCode'],
                                            "fullName": user_data["FullName"],
                                            "email": user_data["Email"],
                                            "faculty": user_data["Faculty"]}
        
        data_combined.append([user_data['UserCode'], str(AId) , str(user_data['UAStatus'])])
```
