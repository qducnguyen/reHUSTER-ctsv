# Dirty EDA

```python
usr_data[123]
```

```python
# dict: AId first
usr_data_clean = {}
for sample in usr_data:
    if sample["AId"] not in not_exists_event and sample["AId"] not in usr_data_clean:
        usr_data_clean[sample["AId"]] = sample

len(usr_data_clean) # 8900 - 4528
```

```python
list(usr_data_clean.keys())[-5:]
```

```python
usr_data_clean[8895]
```

```python
# dict: AId first
act_data_clean = {}
for sample in act_data:
    if sample["AId"] not in not_exists_event and sample["AId"] not in act_data_clean:
        act_data_clean[sample["AId"]] = sample

len(act_data_clean) # 8900 - 4528
```

```python
act_data_clean[8895]
```

```python
data_combined = []

for AId, sample in usr_data_clean.items():
    CreateDate = act_data_clean[AId]['Activities'][0]['CreateDate']
    for user_data in sample["UserActivityLst"]:
        data_combined.append([user_data['UserCode'], str(AId) , str(user_data['UAStatus']), CreateDate])
```

```python
len(data_combined), data_combined[0]
```

```python
df = pd.DataFrame(data_combined,
                  columns=["student_id", "activity_id", "accepted_status", "date"])
df = df.sample(frac = 1).reset_index()
```

```python
df
```

```python
with open("output/dirty_eda_result.csv", "w") as f:
    df.to_csv(f, index=False)
```
